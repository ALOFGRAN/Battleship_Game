```python
from random import randint

board = []
#Draw the board
for x in range(0, 5):
  board.append(["O"] * 5)

def print_board(board):
  for row in board:
    print (" ".join(row))

print_board(board)

#determine ship placement
def random_row(board):
  return randint(0, len(board) - 1)
def random_col(board):
  return randint(0, len(board[0]) - 1)
#assign ship placement to variables
ship_row = random_row(board)
ship_col = random_col(board)

#print the following lines to show ship location in game (for testing purposes only)
#print (ship_row)
#print (ship_col)


for turn in range(4):
  print ("You have ", turn + 1, " turns to guess the ship's location")
  guess_row = int(input("Guess Row: "))
  guess_col = int(input("Guess Col: "))
  #handle a hit ship
  if guess_row == ship_row and guess_col == ship_col:
    print ("Congratulations! You sank my battleship!")
    #end the game due to a sunk ship
    break
  else:
    #to handle guesses on off-board coordinates
    if guess_row not in range(5) or guess_col not in range(5):
      print ("Oops, that's not even in the ocean.")
    #to handle duplicate location guesses
    elif board[guess_row][guess_col] == "X":
      print("You guessed that one already.")
    #to handle a 'miss' guess
    else:
      print ("You missed my battleship!")
      board[guess_row][guess_col] = "X"
    #handle end of game
    if (turn == 3):
      print ("Game Over")
    print_board(board)
```
